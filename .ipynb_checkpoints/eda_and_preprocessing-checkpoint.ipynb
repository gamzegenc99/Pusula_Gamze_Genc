<h3>Tarih Sütunlarının Dönüştürülmesi</h3>

```python
# İlaç kullanım süresini hesaplama
df['Ilac_Baslangic_Tarihi'] = pd.to_datetime(df['Ilac_Baslangic_Tarihi'])
df['Ilac_Bitis_Tarihi'] = pd.to_datetime(df['Ilac_Bitis_Tarihi'])
df['Ilac_Kullanım_Suresi'] = (df['Ilac_Bitis_Tarihi'] - df['Ilac_Baslangic_Tarihi']).dt.days

# Yan etkinin ortaya çıkma süresini hesaplama
df['Yan_Etki_Ortaya_Cikma_Suresi'] = (pd.to_datetime(df['Yan_Etki_Bildirim_Tarihi']) - pd.to_datetime(df['Ilac_Baslangic_Tarihi'])).dt.days

# Orijinal tarih sütunlarını, Uyruk ve Kullanıcı Idsini silebiliriz
df.drop(['Dogum_Tarihi', 'Ilac_Baslangic_Tarihi', 'Ilac_Bitis_Tarihi', 'Yan_Etki_Bildirim_Tarihi', 'Uyruk', 'Kullanici_id'], axis=1, inplace=True)
```

<h3>Eksik Verileri İşleme Yöntemi belirlemek için güncel veri seti üzerinde analiz</h3>

<h4><em>Kategorik değişkenlerin - Yan etki üzerinde istatistiksel etkisi</h4>

```python
categorical_vars = ['Cinsiyet', 'Il', 'Ilac_Adi', 'Alerjilerim', 
                    'Kronik Hastaliklarim', 'Baba Kronik Hastaliklari', 
                    'Anne Kronik Hastaliklari', 'Kiz Kardes Kronik Hastaliklari', 
                    'Erkek Kardes Kronik Hastaliklari', 'Kan Grubu']

# Her bir kategorik değişken için ki-kare testi
for var in categorical_vars:
    # Çapraz tablo oluştur
    cross_tab = pd.crosstab(df[var], df['Yan_Etki'])
    
    # Ki-kare 
    chi2, p, dof, expected = chi2_contingency(cross_tab)
    
    # Sonuçlar
    print(f"Değişken: {var}")
    print(f"Ki-kare Değeri: {chi2:.2f}, p-değeri: {p:.4f}")
    print(f"Serbestlik Derecesi: {dof}")
    print("\n")
```

<h4><em>Sayısal değişkenlerin, korelasyon ile birbirleri arasında nasıl bir ilişki içerisinde olduğuna bakalım:</h4>

```python
numeric_cols = df[['Kilo', 'Boy', 'Yas', 'Ilac_Kullanım_Suresi', 'Yan_Etki_Ortaya_Cikma_Suresi']]

# Korelasyon matrisini oluştur
correlation_matrix = numeric_cols.corr()

# Korelasyon matrisini görselleştirilmesi
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar=True)
plt.title('Sayısal Sütunlar Arası Korelasyon Matrisi')
plt.show()
```

<h4><em>Kategorik değişkenlerin birbiriyleriyle olan ilişkileri</h4>

```python
# Kategorik değişkenler
categorical_vars = ['Cinsiyet', 'Il', 'Ilac_Adi', 'Alerjilerim', 
                    'Kronik Hastaliklarim', 'Baba Kronik Hastaliklari', 
                    'Anne Kronik Hastaliklari', 'Kiz Kardes Kronik Hastaliklari', 
                    'Erkek Kardes Kronik Hastaliklari', 'Kan Grubu']

# Her bir kategorik değişken için diğerleriyle olan ilişkiyi test edelim
for i in range(len(categorical_vars)):
    for j in range(i + 1, len(categorical_vars)):
        var1 = categorical_vars[i]
        var2 = categorical_vars[j]
        
        # Çapraz tablo
        cross_tab = pd.crosstab(df[var1], df[var2])
        
        # Ki-kare testi
        chi2, p, dof, expected = chi2_contingency(cross_tab)
        
        # Sonuçları
        print(f"{var1} ve {var2} arasındaki ilişki:")
        print(f"Ki-kare Değeri: {chi2:.2f}, p-değeri: {p:.4f}")
        print(f"Serbestlik Derecesi: {dof}\n")
```

<h3>Eksik verileri tamamlama</h3>

```python
# Sayısal veriler: Kilo ve Boy için mean ile doldurma 
mean_imputer = SimpleImputer(strategy='mean') 
df[['Kilo', 'Boy']] = mean_imputer.fit_transform(df[['Kilo', 'Boy']])

#Kategorik veriler: Cinsiyet, Kan Grubu,IL,Alerjiler ve Kronik Hastalıklar (aileyle ilgili olanlar dahil) için eksik değerleri 'Bilinmiyor' ile doldurulması
categorical_columns = ['Alerjilerim', 'Cinsiyet', 'Il', 'Kan Grubu','Kronik Hastaliklarim', 
                       'Anne Kronik Hastaliklari', 'Baba Kronik Hastaliklari', 
                       'Erkek Kardes Kronik Hastaliklari', 'Kiz Kardes Kronik Hastaliklari']
# Boş stringleri NaN ile değiştir
df[categorical_columns] = df[categorical_columns].replace(["", " ", "None"], np.nan)
categorical_imputer = SimpleImputer(strategy='constant', fill_value='Bilinmiyor')
df[categorical_columns] = categorical_imputer.fit_transform(df[categorical_columns])

# Kontrol etmek için güncellenmiş eksik verileri inceleyelim
missing_data_after = df.isnull().sum()

missing_data_after
```

<h3>Vücut Kitle İndeksi (BMI) Hesaplama ve Yan etkiler üzerindeki analizi</h3>

```python
# BMI Hesaplama (Kilo kg, Boy cm cinsindendir)
df['BMI'] = df['Kilo'] / (df['Boy'] / 100) ** 2

# BMI kategorilerine ayırma
conditions = [
    (df['BMI'] < 18.5),
    (df['BMI'] >= 18.5) & (df['BMI'] < 25),
    (df['BMI'] >= 25) & (df['BMI'] < 30),
    (df['BMI'] >= 30)
]
categories = ['Zayıf', 'Normal Kilo', 'Fazla Kilo', 'Obez']
df['BMI_category'] = np.select(conditions, categories, default='Bilinmiyor')

# Yan Etkilerin BMI'ye göre dağılımını inceleme
bmi_side_effects = df.groupby('BMI_category')['Yan_Etki'].value_counts()

# Grafik
plt.figure(figsize=(12, 6)) 
bmi_side_effects.unstack().plot(kind='bar', stacked=True)
plt.title('BMI Kategorilerine Göre Yan Etki Dağılımı')
plt.ylabel('Yan Etki Sayısı')
plt.xticks(rotation=45, ha="right") 
plt.tight_layout() 
plt.legend(title="Yan Etki", loc='lower right', bbox_to_anchor=(1.1, 0.5), ncol=1, fontsize=6)
plt.show()
```

<h3>Kategorik Verilerin Sayısal Hale Getirilmesi</h3>

```python
# Initialize LabelEncoder
label_encoder = LabelEncoder()

# 1. Cinsiyet için Label Encoding
df['Cinsiyet'] = label_encoder.fit_transform(df['Cinsiyet'])

# 2. İl, İlaç Adı, Yan Etki ve Alerjilerim için One-Hot Encoding
# 'drop_first=True' ile ilk kategoriyi düşürerek multicollinearity'i önle
df = pd.get_dummies(df, columns=['Il', 'Ilac_Adi', 'Yan_Etki', 'Alerjilerim'], prefix=['Il', 'Ilac', 'YanEtki', 'Alerjiler'], drop_first=True)

# 3. Kronik Hastalıklar için Multi-Label Binarization
for column in ['Kronik Hastaliklarim', 'Baba Kronik Hastaliklari', 'Anne Kronik Hastaliklari', 'Kiz Kardes Kronik Hastaliklari', 'Erkek Kardes Kronik Hastaliklari']:
    binary_columns = df[column].str.get_dummies(sep=', ')
    binary_columns.columns = [f"{column}_{col}" for col in binary_columns.columns]  # Prefix for clarity
    df = pd.concat([df, binary_columns], axis=1)

# 4. Kan Grubu için Label Encoding
df['Kan Grubu'] = label_encoder.fit_transform(df['Kan Grubu'])

# 5. BMI_category için Ordinal Encoding
bmi_mapping = {'Zayıf': 0, 'Normal Kilo': 1, 'Fazla Kilo': 2, 'Obez': 3}
df['BMI_category'] = df['BMI_category'].map(bmi_mapping)

# 6. Orijinal sütunları çıkar
df.drop(columns=['Kronik Hastaliklarim', 'Baba Kronik Hastaliklari', 'Anne Kronik Hastaliklari', 'Kiz Kardes Kronik Hastaliklari', 'Erkek Kardes Kronik Hastaliklari'], inplace=True)

# Sonucu görüntüleyin
print(df.head())
```

<h3>Boyut Düşürme (Dimensionality Reduction) </h3>

<strong> Birleşen Temel Analizi (PCA)</strong>

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

X = df.select_dtypes(include=['float64', 'int64']) 
pca = PCA().fit(X)  # X, sayısal veri

plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_.cumsum())
plt.xlabel('Bileşen Sayısı')
plt.ylabel('Kümülatif Varyans Oranı')
plt.title('Bileşen Sayısına Göre Kümülatif Varyans')
plt.grid()
plt.show()


# Varyans oranlarını ve bileşenleri inceleyin
print("Varyans Oranları:", pca.explained_variance_ratio_)
print("PCA Bileşenleri:", pca.components_)
```

```python
# Kümülatif varyans oranını hesapla
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)

threshold = 0.95
n_components = np.argmax(cumulative_variance >= threshold) + 1  # %95 varyans için gereken bileşen sayısı
print(f"Seçilen Bileşen Sayısı: {n_components}")

# PCA uygulamak için veri setini yeniden oluştur
pca = PCA(n_components=n_components)
X_reduced = pca.fit_transform(X)

# Yeni veri setini kontrol et
print("Yeni Veri Seti (Küçültülmüş):", X_reduced)
print("Yeni Veri Seti (Küçültülmüş) Boyutu:", X_reduced.shape)
```
